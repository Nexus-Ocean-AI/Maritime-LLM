```python
import json
papers_count = 0
count_with_full_text = 0
try:
    
    with open('parsed_papers.jsonl', 'r') as f:
        for line in f:
            data = json.loads(line)
            papers_count += 1
            if 'full_text' in data and data['full_text'] and data['full_text'].strip():
                count_with_full_text += 1
        
    print(f"Number of PDFs with full_text: {count_with_full_text}")
    print(f"Total number of PDFs: {papers_count}")
except FileNotFoundError:
    print("Error: parsed_papers.jsonl not found.")
except json.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")
```
